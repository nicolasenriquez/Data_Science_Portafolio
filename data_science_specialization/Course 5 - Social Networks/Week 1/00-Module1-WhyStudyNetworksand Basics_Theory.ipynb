## Network Definitions and Vocabulary

Network or Graph it's a representation of connections among a set of items.
- The items are called nodes or vertices
- The connections are called edges, link or ties

And we're going to use NetworkX in Python in order to work with some of the networks that we look at. So the first thing you need to know is **how to create an undirected network in NetworkX**.

For example here,

```python
import networkx as nx

# Set the undirected graph object
# Undirected, meaning the edges don't have any direction, or that the direction of the edge is not relevant
G = nx.Graph()

# Add a connection or edge between the vertices A and B
G.add_edge('A', 'B')
# Add a connection or edge between the vertices B and C
G.add_edge('B', 'C')```

I'm adding the edge A, B, which is this edge right here. And then I would add the next edge, which would be the edge B, C. And I could continue adding all the other edges. Now notice that I haven't added the nodes themselves. And so when you add the edge, if the nodes that you're adding, the pairs of nodes that you're adding are not in the graph already, NetworkX will automatically add them to the graph.


### Symmetric and Asymetric Networks

Now let's go back to a network that we had seen before, recall the network between people and the edges here represent friendship, marital ties and family ties among 2,200 people. When you look at this network one thing that you can see is that **these edges are mostly symmetric relationships**. And by that I mean if **A is a friend of B, then B is also a friend of A**. Or at least most of the time, that's the case. And so this network has symmetric relationships.

But that doesn't always have to be the case. So if you look at the food chain network, which represents what animals eat other animals, these **relationships are very asymmetric**. For example, it's very different if you have an edge pointing from the fish to the eagle. That says that the eagle eats the fish, rather than the other way around. So the direction of the edge in this network has a very important meaning to what the edge is trying to represent. 

This suggests that we need **at least two different types of networks**. Some that are **undirected, meaning the edges don't have any direction**, or that the direction of the edge is not really important, and we call these undirected edges. And in NetworkX we use the class `Graph()` to represent these types of networks.

But if we wanted to have a **network that actually has direction**, like the food web example, then you would need directed edges. And to use NetworkX to represent the network like this, we would use the class `DiGraph()`, which is directed graph. And then we would add the edges. 


```python
import networkx as nx

# Set the undirected graph object
# Directed, meaning the edges have direction, or that the direction of the edge is important
G = nx.DiGraph()

# Add a edge between vertices, starting at B, and ending pointing to A
G.add_edge('B', 'A') # (start, end) B -> A

# Add a edge between vertices, starting at B, and ending pointing to C
G.add_edge('B', 'C') # (start, end) B -> C
```

So we will be adding the edge B, A and then the edge B, C. And notice that now the order matters. So it's very different to add the edge B, C than adding the edge C, B, because now these have directions.




### Weighted Networks

The next type of network we're going to look at is the weighted network. The intuition here is that in some networks, some edges carry different weight than others. And so we want a way of capturing this. Let's start with an example of such a network. Suppose a network where the nodes are people and then the edges represent how many times they had lunch together. These are coworkers. And so if you look for example at edge A, B, A and B had lunch together 6 times. Whereas the edge C, E, which says that the nodes C and E had lunch together 25 times. Those are two very different relationships, and so we want a way of capturing this weight. And so **weighted network is a network where edges are assigned a particular weight, typically a numerical weight, that has some meaning in the relationship between the two nodes**.

In NetworkX, we can represent these types of networks also by using the class `Graph(weight=)`. But what we're going to do is when we add an edge, for example the edge A, B, we're going to add an attribute weight, which will have the value 6. And this will allow us to capture these weights on the graph, and the same thing with the edge B, C, which has a weight of 13, and so on.

```python
import networkx as nx

# Set the undirected graph object
# Undirected, meaning the edges don't have any direction
G = nx.Graph()

# Add a edge between the vertices A and B with weight = 6
G.add_edge('A', 'B', weight=6)
# Add a edge between the vertices B and C with weight = 13
G.add_edge('B', 'C', weight=13)
```


### Signed Networks

The next type of network we're going to look at is the signed network. Now, **some networks can carry information about friendship but also about antagonism between people, which could happened due to conflict or disagreement**. So, for example, the websites Epinions and Slashdot **give people the ability to declare not only who their friends are, but also who their enemies are**.

So now let's look at an example of a network that represents who are friends with each other and who are enemies with each other. We'd like to have a way of capturing this. And this is called a signed network, and now **the edges, instead of having a weight, they have a positive or negative sign that represents whether the people are friends or enemies**.

We can construct these types of networks in NetworkX using the class `Graph(sign=)`. And now instead of adding an attribute weight to the edge, we're going to add an attribute which we call sign. And so the edge A, B will have a positive sign. And then the edge B, C will have a negative sign.


```python
import networkx as nx

# Set the undirected graph object
# Undirected, meaning the edges don't have any direction
G = nx.Graph()

# Add a positive (+) connection between the vertices A and B
G.add_edge('A', 'B', sign='+')
# Add a negative (-) connection between the vertices B and C
G.add_edge('B', 'C', weight='-')
```


### Other Edge Attributes: Relationships

So edges can have weights, they can have signs, but they can have a bunch of other things when you think about it. So here's another example. 

Now suppose that the edges are colored, and they're **colored to represent a particular type of relationship between two nodes**. So here we have nodes who are family members or who are friends or who are coworkers or who are neighbors. So in order to represent something like this as you might imagine, you can just use another type of attribute when you add the edge. In this case we call it relation, so when we add the edge A, B, because it has a blue color here, that means they're friends, then we're going to add the attribute relation and we're going to give it the value friend, and the same thing with coworker relationship between B and C. And then the family relationship between D and E. 

So in general, we can have **any type of attribute when we add these edges, we can give the edge any type of attribute that we want**.


```python
import networkx as nx

# Set the undirected graph object
# Undirected, meaning the edges don't have any direction
G = nx.Graph()

# Add a friend relationship edge between the vertices A and B
G.add_edge('A', 'B', relation='friends')
# Add a coworker relationship edge between the vertices B and C
G.add_edge('B', 'C', relation='coworker')
# Add a family relationship edge between the vertices D and E
G.add_edge('D', 'E', relation='family')
# Add a neighbor relationship edge between the vertices E and I

```


### Multigraphs

So the next type of network we're going to look at are called multigraphs. **Intuition for multigraphs is that for a single pair of of nodes, there is no reason why they shouldn't be able to have many different relationships at the same time**.

So let's go back to the example we were looking at. **In this graph, the edges represent some type of relationship between the nodes, which could be family, friend, coworker, or neighbor. There is nothing that stops any two nodes from having two kinds of relationships at the same time**. 

So, for example, this network, in general, could look more like this. Where now the relationship between nodes A and B is not only that they're friends, but they're also neighbors. Or the relationship between B and C is not only that they're family members, but they're also coworkers, and so on. And so this is called a multigraph. It's a network where multiple edges can connect the same pair of nodes, this are also called parallel edges.

In NetworkX, we're going to use a different class in order to represent multigraphs, which is the `MultiGraph()` class. And what we're going to do is when we add an edge to a multigraph, and we're going to add it and give it a particular attribute. And if we want to add a second edge with a different attribute, we can add it again and give it a different attribute. And in fact, you don't have to even give it attributes to a multigraph. 

You can just simply add multiple parallel edges. And then the multigraph class will capture how many edges there are between a given pair of nodes. In this case, we are looking at different relationships, so we are using the attributes, and we would add the rest of the edges.


```python
import networkx as nx

# Set the undirected MultiGraph object (pair of nodes can have multiple edges)
# Undirected, meaning the edges don't have any direction
G = nx.MultiGraph()

# Add a friend relationship edge between the vertices A and B
G.add_edge('A', 'B', relation='friends')
# Add a neighbor relationship edge between the vertices A and B
G.add_edge('A', 'B', relation='neighbor')

# Add a coworker relationship edge between the vertices B and C
G.add_edge('B', 'C', relation='coworker')
# Add a friend relationship edge between the vertices B and C
G.add_edge('B', 'C', relation='family')
```


### Summary

So in summary, we have different types of networks, we saw the following types of networks:
- **Undirected Networks** or undirected graphs, which we use to describe that the edges don't have any direction, or that the direction of the edge is not really important, we use the `Graph()` class in order to represent them


- **Directed Graphs** are used to describe that the edges does have a direction, or that the direction of the edge is important, for which we use the `DiGraph()` class


- **Weighted Networks** is a network where edges are assigned a particular weight, typically a numerical weight, that has some meaning in the relationship between the two nodes, for which we add a weight attribute in the edges when we add the edges in the `Graph(weight=13)`


- **Sign Networks** in the edges, instead of having a weight, they have a positive or negative sign that represents whether the people are friends or enemies, for which we use the `Graph(sign='+' | sign='-')` class but we assign attribute to the edges when we add them


- **Multigraphs** Intuition for multigraphs is that for a single pair of of nodes, there is no reason why they shouldn't be able to have many different relationships at the same time, and for those we use the `MultiGraph()` class, and those can handle multiple edges for the same / each pair of nodes

## Bipartite Graphs

A graph is considered a **bipartite graph** if it has **two sets of nodes which we call L and R, and every single edge connects a node from L to R. So, no edge connects a node from L to another node in L, and no edge connects a node in R to another node in R**.

A good example of this would be that L Nodes represents the fans or the followers for the different football clubs from the first division, and the R Nodes represents all the football clubs from the first division, an so, every one of the fans ($l_{i}\text{, }\forall l_{i} \in L$) is connected to a specific team in R ($r_{i}\text{, }\forall r_{i} \in R$), and as we can see, no fan is related to one another, nor a football club is linked to another club, and the only connection exists between the fans and the football club that they like.

NetworkX does not have a separate class for bipartite graph, but it does have a set of algorithms that allow us to study, analyze and to do things with them, this is how we would approach the problem applying code

```python
import networkx as nx
from networkx.algorithms import bipartite

# Step 1: Set the Graph object
# There's no separate class for bipartite graphs
B = nx.Graph()

# Step 2: Add the Nodes fron the left side of the bipartite groups represented by 0 (in the examples: fans)
B.add_nodes_from(['A', 'B', 'C', 'D', 'E'], bipartite=0)  # labels for the left side / group 0 of bipartite

# Step 3: Add the Nodes fron the right side of the bipartite groups represented by 1 (in the examples: fc clubs)
B.add_nodes_from([1, 2, 3, 4], bipartite=0)  # labels for the right side / group 1 of bipartite

# Step 4: Add the Edges for all the Nodes
B.add_edges_from([('A', 1), ('B', 1), ('C', 1), ('C', 3),
                  ('D', 2), ('E', 3), ('E', 4)])


# TIP Nº1
# Check if a Graph is Bipartite
In: bipartite.is_bipartite(B)  # Check if B is bipartite
Out: True

# Note: If we add a new edge we would break the rule and the Graph B would no longer be bipartite
In: B.add_edge('A', 'B')       # Add new connection between nodes A and B
In: bipartite.is_bipartite(B)  # Check if B is bipartite
Out: False                     # This happends because the new edge connection breaks the bipartite rule


# TIP Nº2
# Check if a Subset of Nodes in a Graph is Bipartite
In: X = set([1, 2, 3, 4])            # Create the subset of elements
In: bipartite.is_bipartite(B, X)     # Check if it's bipartite
Out: True

In: X = set(['A', 'B', 'C', 'D', 'E'])
In: bipartite.is_bipartite(B, X)
Out: True

In: X = set([1, 2, 3, 4, 'A'])
In: bipartite.is_bipartite(B, X)
Out: False


# TIP Nº3
# How to Detect Which Sets of Nodes of a Bipartite Graph if we don't know them or if even exists
In: bipartite.sets(B)     # Estimate the bipartite sets
Out: (('A', 'B', 'C', 'D', 'E'), (1, 2, 3, 4)) 

# Note: If we add a new edge between nodes A and B, we break the bipartite graph rule
In: B.add_edge('A', 'B')  # Add edge between nodes (A,B)
In: bipartite.sets(B)     # Estimate the bipartite sets
Out: 'NetworkXError: Graph is not bipartite'
```


### Projected Graphs

Imagine we have the same example as before, the fans nodes and the football club sets. Now imagine that you were interested in creating a network among the fans, **and have the edges represent some type of affinity in terms of what teams they follow**. Whether they kind of tend to follow the same teams or not. **This kind of network could be important for viral marketing**. So, **if two people tend to follow the same teams, they may also like the same other type of product**, and so you would be interested in knowing who is likely to impact whom in terms of that other product, and the fact that they follow the same kinds of teams might give you that hint, and so **these this kind of network could be useful for certain things**.

- **L-Bipartite Graph Projection: ** it's a network among the nodes in one side of the group where each pair of nodes is connected if they have a common neighbor in the R side of the bipartite graph. So, in this case, there would be exactly the **network between the fans, such that they're connected if they have at least one team in common**. You would have a similar definition for the R-bipartite graphs, so that would be a network among the football clubs, and **two teams will be connected if they have at least one fan in common**.


We can estimate this projected network by using NetworkX, and the way to do it is to define the graph and add all the edges. Then, we defined the set X to be the set of fans, and then, we'd use the function projected graph, and it takes the input B, which is the bipartite graph, and then X which is the set of nodes that you want the projected graph on, and then you get the network. 

```python
import networkx as nx
from networkx.algorithms import bipartite

# Set the Graph object
# There's no separate class for bipartite graphs
B = nx.Graph()

# Add the Edges
B.add_edges_from([('A', 1), ('B', 1), ('C', 1), ('D', 1), ('H', 1),
                  ('B', 2), ('C', 2), ('D', 2), ('E', 2), ('G', 2),
                  ('E', 3), ('F', 3), ('H', 3), ('J', 3),
                  ('E', 4), ('I', 4), ('J', 4)])
                  
# Define the set
X = set(['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'])

# Estimate the projected graph or network of fans that have 1 club in common
P = bipartite.projected_graph(B, X)    
```

What if you wanted the projected graph for the teams? 

So in this case you would now say X is the set of basketball teams, and then P will be the projected graph using that set, and then this is the network that would come out. So in this network, the nodes 1 and 2 are connected because they both have at least one fan in common.

```python

### How to Read File Format Nº1: Adjacency List

`G_adjlist.txt` is the adjaceny list representation of G1.

The adjacency list format is useful for graphs without nodes or edge attributes. This format consists of lines with node labels. The first label in each line is the source node. All subsequent labels correspond to connected target nodes.

It can be read as follows:
* `0 1 2 3 5` $\rightarrow$ node `0` is adjacent to nodes `1, 2, 3, 5`
* `1 3 6` $\rightarrow$ node `1` is (also) adjacent to nodes `3, 6`
* `2` $\rightarrow$ node `2` is (also) adjacent to no new nodes
* `3 4` $\rightarrow$ node `3` is (also) adjacent to node `4` 

Note that adjacencies are only accounted for once (e.g. node `2` is adjacent to node `0`, but node `0` is not listed in node `2`'s row, because that edge has already been accounted for in node `0`'s row).

For example, looking at the first line of G_adjlist.txt, we can see nodes 1, 2, 3, and 5 are all connected to node 0. Note that each line only adds nodes that weren't in previous lines as source nodes. For instance, in the second line of G_adjlist.txt, node 0 is not included because the connection between node 0 and 1 has already been accounted for.

### Chess Example

Now let's load in a more complex graph and perform some basic analysis on it. We will be looking at chess_graph.txt, which is a directed graph of chess games in edge list format.

````python
!head -5 chess_graph.txt
````

Each node is a chess player, and each edge represents a game. The first column with an outgoing edge corresponds to the white player, the second column with an incoming edge corresponds to the black player.

The third column, the weight of the edge, corresponds to the outcome of the game. A weight of 1 indicates white won, a 0 indicates a draw, and a -1 indicates black won.

The fourth column corresponds to approximate timestamps of when the game was played.

We can read in the chess graph using `read_edgelist`, and tell it to create the graph using a `nx.MultiDiGraph`.

````python
chess = nx.read_edgelist('chess_graph.txt', data=[('outcome', int), ('timestamp', float)], 
                         create_using=nx.MultiDiGraph())
````

````python
# Check if the Graph is: directed?, multigraph?

print(f"Is the Graph Directed? - {chess.is_directed()}\nIs it a MultiGraph? - {chess.is_multigraph()}")
````

````python
# Check if we loaded the data correctly
chess.edges(data=True)
````

Looking at the degree of each node, we can see how many games each person played. A dictionary is returned where each key is the player, and each value is the number of games played.

````python
# Estimate the number of games for every node
games_played = chess.degree()
games_played
````

Using list comprehension, we can find which player played the most games.

````python
max_value = max(games_played.values())
max_key, = [i for i in games_played.keys() if games_played[i] == max_value]

print('player {}\n{} games'.format(max_key, max_value))
````

Let's use pandas to find out which players won the most games. First let's convert our graph to a DataFrame.

````python
df = pd.DataFrame(chess.edges(data=True), columns=['white', 'black', 'outcome'])
df.head()
````

Next we can use a lambda to pull out the outcome from the attributes dictionary.

````python
df['outcome'] = df['outcome'].map(lambda x: x['outcome'])
df.head()
````

To count the number of times a player won as white, we find the rows where the outcome was '1', group by the white player, and sum.

To count the number of times a player won as back, we find the rows where the outcome was '-1', group by the black player, sum, and multiply by -1.

The we can add these together with a fill value of 0 for those players that only played as either black or white.

````python
won_as_white = df[df['outcome']==1].groupby('white').sum()
won_as_black = -df[df['outcome']==-1].groupby('black').sum()
win_count = won_as_white.add(won_as_black, fill_value=0)
win_count.head()
````

Using `nlargest` we find that player 330 won the most games at 109.

````python
win_count.nlargest(5, 'outcome')
````
